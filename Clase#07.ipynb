```python
a = np.array([[1,2],[3,4]])
b = np.array([5,6])
```

```python
a.ndim
```

```python
b.ndim
```

```python
np.concatenate((a,b),axis=0) #dara error por diferencia de dimensiones
```

```python
b = np.expand_dims(b,axis=0) #para expandir, igualar dimensiones y poder concatenar
```

```python
np.concatenate((a,b),axis=0)
```

```python
np.concatenate((a,b),axis=1) #dara error por diferencia de dimensiones
```

```python
np.concatenate((a,b.T),axis=1) #aplica transpuesta para permitir concatenar
```
