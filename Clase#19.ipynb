```python
import pandas as pd
import numpy as np
```

```python
#Concat
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
'B':['B0','B1','B2','B3'],
'C':['C0','C1','C2','C3'],
'D':['D0','D1','D2','D3']})

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
'B':['B4','B5','B6','B7'],
'C':['C4','C5','C6','C7'],
'D':['D4','D5','D6','D7']})
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
#Merge
izq = pd.DataFrame({'key':['k0','k1','k2','k3',],
'A':['A0','A1','A2','A3'],
'B':['B0','B1','B2','B3']})
izq
```

```python
der = pd.DataFrame({'key':['k0','k1','k2','k3',],
'C':['C0','C1','C2','C3'],
'D':['D0','D1','D2','D3']})
der
```

```python
izq.merge(der)
```

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3',],
'A':['A0','A1','A2','A3'],
'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2','k3',],
'C':['C0','C1','C2','C3'],
'D':['D0','D1','D2','D3']})

izq.merge(der, on='key') #Dará error porque key != key_2
```

```python
izq.merge(der, left_on='key',right_on='key_2') #Así no nos dará el error de arriba
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3',],
'A':['A0','A1','A2','A3'],
'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2',np.nan,],
'C':['C0','C1','C2','C3'],
'D':['D0','D1','D2','D3']})

izq.merge(der, left_on='key',right_on='key_2')
```

```python
izq.merge(der, left_on='key',right_on='key_2',how='left')
```

```python
izq.merge(der, left_on='key',right_on='key_2',how='inner')
```

```python
izq.merge(der, left_on='key',right_on='key_2',how='right')
```

```python
izq = pd.DataFrame({'A':['A0','A1','A2'],
'B':['B0','B1','B2']},
index=['k0','k1','k2'])

der = pd.DataFrame({'C':['C0','C1','C2'],
'D':['D0','D1','D2']},
index=['k0','k2','k3'])

izq.join(der)
```

```python
izq.join(der,how='inner')
```

```python
izq.join(der,how='left')
```

```python
izq.join(der,how='right')
```

```python
izq.join(der,how='outer')
```
