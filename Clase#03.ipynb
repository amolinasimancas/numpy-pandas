```python
import numpy as np
arr = np.array([1,2,3,4])
arr.dtype #tipo de dato del array
```

```python
arr = np.array([1,2,3,4], dtype='float64')
arr.dtype
arr #se evidencia el cambio a decimal
```

```python
arr = np.array([1,2,3,4])
arr = arr.astype(np.float64)
arr #cambio a float
```

```python
arr = np.array([0,1,2,3,4])
arr = arr.astype(np.bool_)
arr #cambio a bool
```

```python
arr = np.array([0,1,2,3,4])
arr = arr.astype(np.string_)
arr #cambio a string
```

```python
arr = np.array(['0','1','2','3','4'])
arr = arr.astype(np.int8)
arr #cambio a integer
```
